## Lab 1 (Image brightening)

Images from your lab always come out a bit darker. You have been aske to use image
processing to automate brightening of photos.
Write a simple script that can
- Load an image
- Increase image brightness
- Display old and new image

Libraries needed
- Pillow
- Matplotlib

```python
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt
```

Loading the image

```python
im = Image.open("dark-car.jpg")
```

Using the image brightness enhancer and setting the factor to 1.8 to increase the brightness of the image


```python
enhancer = ImageEnhance.Brightness(im)
factor = 1.8 #brightens the image
im_output = enhancer.enhance(factor)
im_output.save('brightened-image.png')
```

showing  images side by side

```python
im2 = Image.open('brightened-image.png')
f = plt.figure()
f.add_subplot(1,2, 1)
plt.xlabel("Original Image")
plt.imshow(im)
f.add_subplot(1,2, 2)
plt.xlabel("brightened Image")
plt.imshow(im2)
plt.show(block=True)
```
